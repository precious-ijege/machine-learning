```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
column_names = ["Population","Profit"]  
data = pd.read_csv('ex1data1.txt', names = column_names)
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
# Split the dataset into the training set and test set
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
# Creating a LinearRegression object
linearRegressor = LinearRegression()
```

```python
# Fitting in our training set and training the model
linearRegressor.fit(xTrain, yTrain)
```

```python
# Predicting the test set results
yPrediction = linearRegressor.predict(xTest)
```

```python
# Visualising the training set results
plt.scatter(xTrain, yTrain, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```

```python
# Visualising the test set results
plt.scatter(xTest, yTest, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.show()
```
